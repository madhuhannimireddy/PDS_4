```python
#set a seed (to ensure work reproducibility) and take a random sample of 25 observations and find the mean Glucose and highest Glucose values of this sample and compare these statisticswith the population statistics of the same variable. You should use charts for this comparison.
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt



np.random.seed(123)


df = pd.read_csv("diabetes.csv")


sample = df.sample(n=25)


sample_mean = sample['Glucose'].mean()
sample_max = sample['Glucose'].max()


pop_mean = df['Glucose'].mean()
pop_max = df['Glucose'].max()


print("Sample Mean Glucose: ", sample_mean)
print("Sample Highest Glucose: ", sample_max)
print("Population Mean Glucose: ", pop_mean)
print("Population Highest Glucose: ", pop_max)


plt.bar(['Sample', 'Population'], [sample_mean, pop_mean])
plt.title('Mean Glucose')
plt.show()


plt.bar(['Sample', 'Population'], [sample_max, pop_max])
plt.title('Highest Glucose')
plt.show()
```

```python
#Find the 98th percentile of BMI of your sample and the population and compare the results using charts
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt


np.random.seed(123)
sample_bmi = np.random.normal(25, 2, size=1000)


sample_98th = np.percentile(sample_bmi, 98)


pop_bmi = np.random.normal(25, 2, size=100000)


pop_98th = np.percentile(pop_bmi, 98)


plt.hist(sample_bmi, bins=30)
plt.axvline(sample_98th, color='r', linestyle='--', label='98th percentile')
plt.legend()
plt.title('Sample BMI Distribution')
plt.xlabel('BMI')
plt.ylabel('Frequency')
plt.show()


plt.hist(pop_bmi, bins=30)
plt.axvline(pop_98th, color='r', linestyle='--', label='98th percentile')
plt.legend()
plt.title('Population BMI Distribution')
plt.xlabel('BMI')
plt.ylabel('Frequency')
plt.show()
```

```python
#Using bootstrap (replace= True), create 500 samples (of 150 observation each) from the population and find the average mean, standard deviation and percentile for BloodPressure and compare this with these statistics from the population for the same variable. Again, you shouldcreate charts for this comparison
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt


np.random.seed(123)


data = pd.read_csv('diabetes.csv')


bp = data['BloodPressure']


pop_mean = bp.mean()
pop_std = bp.std()
pop_perc = np.percentile(bp, 90)


bootstrap_stats = []


for i in range(500):
    sample = np.random.choice(bp, size=150, replace=True)
    sample_mean = sample.mean()
    sample_std = sample.std()
    sample_perc = np.percentile(sample, 90)
    bootstrap_stats.append((sample_mean, sample_std, sample_perc))


bootstrap_stats = pd.DataFrame(bootstrap_stats, columns=['Mean', 'Std', '90th Perc'])


plt.boxplot([bootstrap_stats['Mean'], [pop_mean]])
plt.xticks([1, 2], ['Bootstrap', 'Population'])
plt.title('Comparison of Mean BloodPressure')
plt.ylabel('BloodPressure')
plt.show()


plt.boxplot([bootstrap_stats['Std'], [pop_std]])
plt.xticks([1, 2], ['Bootstrap', 'Population'])
plt.title('Comparison of BloodPressure Standard Deviation')
plt.ylabel('BloodPressure')
plt.show()


plt.boxplot([bootstrap_stats['90th Perc'], [pop_perc]])
plt.xticks([1, 2], ['Bootstrap', 'Population'])
plt.title('Comparison of 90th Percentile of BloodPressure')
plt.ylabel('BloodPressure')
plt.show()
```
